# 1. Creación de Entorno de Trabajo

## 1.1 Acceso a Patagón: 
Una parte crucial para comenzar a trabajar en Patagón es configurar adecuadamente el acceso a través de SSH. Sigue estos pasos:

### 1.1.1 Solicitud de Acceso
Primero, debes solicitar acceso a Patagón y asegurarte de tener un usuario y una contraseña válidos proporcionados por los administradores del sistema.

### 1.1.2 Configuración del Cliente SSH
Luego, configura tu cliente SSH para acceder a Patagón de manera eficiente. Esto se puede lograr agregando una entrada en tu archivo de configuración SSH. Abre o crea el archivo ~/.ssh/config en tu dispositivo y agrega lo siguiente

```config
Host *
        AddKeysToAgent yes
Host patagon
        HostName patagon.uach.cl
        Port 2237
        User your_username
        IdentityFile ~/.ssh/your_private_key_filename
```

Reemplaza `your_username` con tu nombre de usuario en Patagón y `your_private_key_filename` con la ubicación de tu clave privada SSH. Esto permitirá un acceso más sencillo y seguro a Patagón.

### 1.1.3 Acceso a Patagón
Una vez configurado, para acceder a Patagón, simplemente abre tu terminal y ejecuta:

```bash
~ ssh patagon
```
## 1.2 Editores de texto e IDEs remotas

En Patagón, tendrás acceso a editores de texto básicos como `nano`. Sin embargo, si prefieres trabajar con un entorno más completo y potente, puedes aprovechar la extensión "SSH Remote" de VSCode desde tu computadora local. Esta extensión ofrece las siguientes ventajas:

- Edición de Código Remoto: Te permite escribir y editar código en tu editor de código preferido (VSCode) directamente en Patagón.

- Transferencia de Archivos: Puedes transferir archivos fácilmente entre tu computadora local y Patagón mediante una interfaz gráfica de arrastrar y soltar.

Para comenzar a usar la extensión "SSH Remote" de VSCode, consulta esta <a href="https://code.visualstudio.com/docs/remote/ssh-tutorial">guía de tutorial de VSCode SSH Remote.</a>

Con esta configuración, tendrás un entorno de trabajo eficiente y flexible en Patagón que se adapta a tus necesidades y preferencias.


## 1.3 Creación de Entorno de Trabajo: 
Para descargar y usar contenedores al usar srun o sbatch debes seguir los siguientes pasos:
### 1.3.1 Autenticación para Descargar Contenedores

En primer lugar, deberás crear un archivo en la ruta `~/.config/enroot/.credentials` en base al siguiente template:


```.credentials
# NVIDIA GPU Cloud (both endpoints are required)
machine nvcr.io login $oauthtoken password APIKEY
machine authn.nvidia.com login $oauthtoken password APIKEY

# DockerHub
# machine auth.docker.io login <login> password <passord>

# Google Container Registry
# machine gcr.io login oauth2accesstoken password $(gcloud auth print-access-token)
# machine gcr.io login _json_key password $(jq -c '.' $GOOGLE_APPLICATION_CREDENTIALS | sed 's/ /\\u0020/g')
```

### 1.3.2 Descargar contenedores desde DockerHub [en progreso]

La descarga de un contenedor se hace al momento de ejecutar una tarea, solo la primera vez, y sigue la estructura

```bash
~ srun --container-name=<nombre-arbitrario> --container-image=<url-imagen-contenedor> <comando-de-ejecucion>

```
Por ejemplo, si quisieramos descargar el contenedor `archlinux` desde DockerHub

```bash
~ srun --container-name=arch --container-image='archlinux:base-devel'

```

Para abrir una sesión interactiva dentro del contenedor e instalar paquetes debes acceder mediante:

```bash
~ srun -p cpu -v --container-name=arch --container-remap-root --pty bash
```

Dentro del contenedor podrás instalar el software necesario como, por ejemplo, Python con la siguiente instrucción: `pacman -S python`

Para mayor información recomiendo visitar la <a href="https://patagon.uach.cl/patagon/tutoriales/administracion-contenedores">documentación de Patagón sobre la administración de contenedores</a>.
